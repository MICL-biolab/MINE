```python
import matplotlib.pyplot as plt
from scipy.ndimage import gaussian_filter1d
```

```python
_log = '/data1/lmh_data/MINE/additional/cycle_error/exp.log'
```

```python
def read_train_info(_path):
    with open(_path, 'r') as f:
        _datas = f.readlines()
    train_acc, test_acc = [], []
    for i in range(3, len(_datas), 3):
        _data = _datas[i].split()
        _train_acc, _test_acc = float(_data[-3]), float(_data[-1])
        train_acc.append(_train_acc)
        test_acc.append(_test_acc)
    return train_acc, test_acc
#     return gaussian_filter1d(train_acc, sigma=6), gaussian_filter1d(test_acc, sigma=3)

train_acc, test_acc = read_train_info(_log)
```

```python
x = [i for i in range(len(train_acc))]
plt.plot(x, train_acc, label='train')
plt.plot(x, test_acc, label='validation')

plt.xlabel('epoch')
plt.ylabel('accuracy')
plt.legend()
plt.savefig('figure_file/cycle-error.pdf', format='pdf', bbox_inches='tight')
plt.show()
```
