```python
import os
import matplotlib.pyplot as plt
from matplotlib_venn import venn2
```

```python
hr_path = "/data1/lmh_data/MMSR_complete/draw/fithic/hr_chr19_1000b/outputs/FitHiC.spline_pass1.res1000.significances.txt"
complete_path = "/data1/lmh_data/MMSR_complete/draw/fithic/replaced_chr19_1000b/outputs/FitHiC.spline_pass1.res1000.significances.txt"
enhanced_path = "/data1/lmh_data/MMSR_complete/draw/fithic/replaced_chr19_result/outputs/FitHiC.spline_pass1.res1000.significances.txt"
low5_path = "/data1/lmh_data/MMSR_complete/draw/fithic/chr19_low5/outputs/FitHiC.spline_pass1.res5000.significances.txt"
low10_path = "/data1/lmh_data/MMSR_complete/draw/fithic/chr19_low10/outputs/FitHiC.spline_pass1.res10000.significances.txt"
```

```python
def get_significances(file_path, cut_off):
    _tuples = []
    with open(file_path, 'r') as file_object:
        data = file_object.readlines()[1:]
        for i in range(len(data)):
            _s = str.split(data[i])
            q_value = float(_s[6])
            if q_value < cut_off:
                _tuples.append((int(_s[1]), int(_s[2])))
    return _tuples

hr_sig = get_significances(hr_path, 6.50223160164796e-10)
complete_sig = get_significances(complete_path, 6.45061963681413e-10)
enhanced_sig = get_significances(enhanced_path, 6.444005214605012e-10)
low5_sig = get_significances(low5_path, 1.6108856435020183e-08)
low10_sig = get_significances(low10_path, 6.43141576813357e-08)
```

```python
venn2(subsets = [set(hr_sig), set(enhanced_sig)], set_labels = ('1kb', 'enhanced'), set_colors=((247/255, 129/255, 191/255), (51/255, 160/255, 44/255)))
plt.savefig('fig/fig6_significant_1kb+enhanced.pdf')
plt.show()
```

```python
venn2(subsets = [set(hr_sig), set(complete_sig)], set_labels = ('1kb', 'complete'), set_colors=((247/255, 129/255, 191/255), (166/255, 206/255, 227/255)))
plt.savefig('fig/fig6_significant_1kb+complete.pdf')
plt.show()
```

```python
venn2(subsets = [set(hr_sig), set(low5_sig)], set_labels = ('1kb', '5kb'), set_colors=((247/255, 129/255, 191/255), (177/255, 89/255, 40/255)))
plt.savefig('fig/fig6_significant_1kb+5kb.pdf')
plt.show()
```
