```python
import os
import copy
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib_venn import venn2
```

```python
GM12878_ATAC_H3K27ac_H3K4me3_2_100_all_enhanced_sig = np.load('../fig2/temp/GM12878_ATAC_H3K27ac_H3K4me3_2_100_all_enhanced_sig.npy', allow_pickle=True).item()
auxiliary_files_path = '/data1/lmh_data/MINE/source/GM12878'
```

```python
def read_bed_file(path):
    _peaks = dict()
    with open(path)as f:
        for line in f:
            datas = line.strip().split()
            left, right = int(int(datas[1])/1000), int(int(datas[2])/1000)
            peaks = list(range(left, right+1))
            if datas[0] not in _peaks.keys():
                _peaks[datas[0]] = set()
            for i in peaks:
                _peaks[datas[0]].add(i)
    return _peaks

def get_hit_indicators(significants, peaks, chr):
    results = set()
    for sig in significants:
        x, y = int(sig[0]), int(sig[1])
        if x in peaks or y in peaks:
            results.add((x, y, chr))
    return results

def show_venn(chia_pet_set, chip_seq_set):
    venn2(subsets = [chip_seq_set, chia_pet_set], set_labels = ('chip_seq', 'chia_pet'), set_colors=((247/255, 129/255, 191/255), (51/255, 160/255, 44/255)))
    plt.show()

def figure(chip_seq_path, chia_pet_path):
    chia_pet_sets = set()
    chip_seq_sets = set()
    chip_seq_peaks = read_bed_file(chip_seq_path)
    chia_pet_peaks = read_bed_file(chia_pet_path)
    for chr in range(1, 23):
        chip_seq_enhanced_result = get_hit_indicators(GM12878_ATAC_H3K27ac_H3K4me3_2_100_all_enhanced_sig[chr], chip_seq_peaks['chr{}'.format(chr)], chr)
        chia_pet_enhanced_result = get_hit_indicators(GM12878_ATAC_H3K27ac_H3K4me3_2_100_all_enhanced_sig[chr], chia_pet_peaks['chr{}'.format(chr)], chr)

        chip_seq_sets = chip_seq_sets | chip_seq_enhanced_result
        chia_pet_sets = chia_pet_sets | chia_pet_enhanced_result
    show_venn(chia_pet_sets, chip_seq_sets)
        
```

```python
chip_seq_path = os.path.join(auxiliary_files_path, 'CTCF_peaks.bed')
chia_pet_path = os.path.join(auxiliary_files_path, 'CTCF_chiapet_peaks.bed')
figure(chip_seq_path, chia_pet_path)
```
